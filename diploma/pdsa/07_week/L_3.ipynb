# üß† Greedy Algorithm ‚Äì Minimizing Maximum Lateness (Scheduling with Deadlines)

---

## üß© Problem Setting

We are given:

* A **single resource** (like a 3D printer).
* `n` **jobs** (or users) want to use this resource.
* Each job `i` has:

  * A required time `T·µ¢` (how long the job takes on the resource),
  * A **deadline** `D·µ¢` (by when it wants to be completed).

### ‚è∞ Objective

Schedule **all jobs** (we *must* schedule all) such that:

> **Maximum Lateness** is minimized
> i.e., minimize
>
> $$
> $$

\max\_i L·µ¢ \text{ where } L·µ¢ = \max(0, F·µ¢ - D·µ¢)
]
Here,

* `F·µ¢` = actual finish time of job `i`
* `L·µ¢` = how late job `i` finishes

We do **not**:

* Remove jobs (unlike in interval scheduling),
* Try to minimize total lateness (`‚àë L·µ¢`), only **max** lateness.

---

## üìå Greedy Algorithm Design

### ‚úÖ Greedy Strategy:

**Schedule jobs in increasing order of deadline (D·µ¢)**

---

## ‚ùå Tempting (But Wrong) Greedy Approaches

### 1. **Shortest Job First (SJF)**

> Sort jobs by increasing `T·µ¢` (time taken)
> Intuition: Quickly finish small jobs to reduce build-up

#### ‚ùóCounterexample:

* Job A: `T=1`, `D=100`
* Job B: `T=10`, `D=10`

SJF order: A ‚Üí B

* A finishes at `1` (on time),
* B finishes at `11`, but `D=10` ‚áí `Lateness = 1`

Correct order: B ‚Üí A

* B finishes at `10` (on time), A finishes at `11` but `D=100` ‚áí `Lateness = 0`

‚úÖ Result: **Better schedule by violating SJF**

---

### 2. **Least Slack Time First**

> Slack = `D·µ¢ - T·µ¢` (how much room there is)
> Schedule jobs with **least slack** first

#### ‚ùóCounterexample:

* Job A: `T=1`, `D=2` ‚áí Slack = 1
* Job B: `T=10`, `D=10` ‚áí Slack = 0

Slack order: B ‚Üí A

* B finishes at 10, A finishes at 11
* A's deadline was 2 ‚áí Lateness = 9

Better order: A ‚Üí B

* A finishes at 1 (on time), B at 11 ‚áí Lateness = 1

‚úÖ Result: **Slack-based ordering fails**

---

## ‚úÖ Correct Greedy Algorithm: **Earliest Deadline First (EDF)**

> Sort jobs by increasing deadlines
> Schedule them in that order

Why might this work?

* You address the **most urgent** jobs first (not necessarily smallest or tightest).

---

## üß™ Proving EDF is Optimal

We use a classic **Exchange Argument**.

---

## üß† Key Observations

### 1. **No Idle Time in Optimal Schedule**

* Any idle time can be ‚Äúcompressed‚Äù without increasing lateness.
* Moving jobs earlier keeps or reduces lateness.
* So, always assume optimal schedule has no gaps.

### 2. **Greedy Schedule = No Inversions**

> An **inversion** occurs if a job `i` appears before job `j`, but `D·µ¢ > D‚±º`

* Greedy schedule: sorted by `D·µ¢`, so **no inversions** possible

---

## üõ† Exchange Argument

### Goal:

Convert **any optimal schedule** into **greedy schedule** using **local swaps** that don‚Äôt increase lateness.

### Steps:

1. Start with any **optimal schedule** `O`, assume:

   * All jobs are scheduled
   * No idle time

2. **If O has an inversion**:

   * There exist jobs `i` and `j` where:

     * `i` comes before `j` in `O`
     * but `D·µ¢ > D‚±º` (inversion)

3. Identify the **first inversion**: it must be between two **adjacent jobs**.

4. **Swap i and j** to fix the inversion.

5. Show that:

   * The **lateness of j** gets better (finishes earlier)
   * The **lateness of i** increases, but not more than the lateness of j before the swap.

   So:

   $$
   \max(L_i, L_j) \text{ remains same or decreases}
   $$

6. Repeat for all inversions. After at most `n(n-1)/2` swaps, no inversions remain ‚áí we get the greedy order.

---

## üßæ Handling Equal Deadlines

* If multiple jobs have the **same deadline**, any permutation of them is valid.
* Greedy schedule picks some arbitrary order.
* Optimal schedule might differ, but **lateness stays the same** since the total block duration is unchanged.

---

## ‚úÖ Final Conclusion

> **The Greedy EDF Algorithm (Earliest Deadline First)** is optimal for minimizing **maximum lateness**.

---

## üìä Algorithm Summary

### **Input:**

Jobs `1..n`, each with:

* `T·µ¢`: Time needed
* `D·µ¢`: Deadline

### **Steps:**

1. Sort jobs by increasing deadline `D·µ¢` ‚Üí O(n log n)

2. Schedule in that order:

   * First job starts at time `0`
   * Each job starts when the previous one finishes

3. Compute for each job:

   * Finish time: `F·µ¢ = S·µ¢ + T·µ¢`
   * Lateness: `L·µ¢ = max(0, F·µ¢ - D·µ¢)`

4. Compute:

   * `max L·µ¢` across all jobs

---

## üß† Time Complexity

* Sorting: O(n log n)
* Scheduling and computing lateness: O(n)
* **Total**: O(n log n)

## üß† Key Learnings

* Greedy algorithms **must be proven** with care. A reasonable-sounding heuristic can **fail badly**.
* Use:

  * **Counterexamples** to disprove wrong heuristics
  * **Exchange arguments** to prove optimality
* Earliest Deadline First is **provably optimal** for minimizing **maximum lateness** in scheduling with deadlines.
