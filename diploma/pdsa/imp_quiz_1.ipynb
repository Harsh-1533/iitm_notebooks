### üîπ **1. Count Loops**

#### üß† Rule:

* Each loop that runs `n` times contributes **O(n)**.
* **Nested loops** multiply their complexities.

#### üîç Examples:

```python
for i in range(n):           # O(n)
    print(i)
```

```python
for i in range(n):           # O(n)
    for j in range(n):       # O(n)
        print(i, j)          # Total = O(n √ó n) = O(n¬≤)
```

```python
for i in range(n):           # O(n)
    for j in range(i):       # O(i) ‚Üí 0 to i-1
        print(i, j)          # Total ‚âà O(n¬≤) (because 1 + 2 + ... + n = n(n+1)/2)
```

---

### üîπ **2. Count Recursive Calls**

Check:

* How many recursive calls are made?
* What happens to the input size in each call?

#### üîç Examples:

```python
def fun(n):
    if n <= 1:
        return 1
    return fun(n-1) + fun(n-1)
```

* Each call spawns **2 calls** ‚Üí exponential ‚Üí **O(2‚Åø)**

```python
def fun(n):
    if n <= 1:
        return 1
    return fun(n-1)
```

* Only 1 call each time ‚Üí depth = n ‚Üí **O(n)**

```python
def fun(L, n):
    for i in range(n):         # O(n)
        print(L[i])
    fun(L, n-1)                # Recursive ‚Üí O(n)
```

* It calls `n` times, each time running a loop of `n, n-1,...` ‚Üí **O(n¬≤)**

---

### üîπ **3. Watch for Logarithmic Behavior**

Any time the input size is **divided or halved**, complexity is **logarithmic**.

#### üîç Examples:

```python
while n > 1:
    n = n // 2         # O(log n)
```

```python
def binarySearch(arr, low, high):
    if low > high:
        return -1
    mid = (low + high) // 2
    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binarySearch(arr, mid + 1, high)
    else:
        return binarySearch(arr, low, mid - 1)
```

* This reduces the search space by half every time ‚Üí **O(log n)**

---

### üîπ **4. Look for Divide-and-Conquer**

This is where:

* Input is split (usually by half)
* Multiple recursive calls are made
* Combine results

#### üîç Example: Merge Sort

```python
def mergeSort(arr):
    if len(arr) > 1:
        mid = len(arr)//2
        L = arr[:mid]
        R = arr[mid:]
        mergeSort(L)
        mergeSort(R)
        merge(L, R, arr)
```

* Split size: log n
* Merge each level: O(n)
* Total: **O(n log n)**

---

### üîπ **5. Drop Constants and Lower Terms**

In Big-O:

* O(2n) ‚Üí O(n)
* O(n + log n) ‚Üí O(n)
* O(5n¬≤ + 3n) ‚Üí O(n¬≤)

| Pattern                          | Time Complexity |

## üß† Practice Tip:

When you see an algorithm, ask yourself:

1. How many times is the main block repeated?
2. Does the input size shrink or stay the same?
3. Are there nested operations?
4. Is recursion creating multiple branches?

Over time, you‚Äôll be able to **instantly spot patterns** like:

* `for i in range(n):` ‚Üí O(n)
* `for i in range(n): for j in range(n):` ‚Üí O(n¬≤)
* `while n > 1: n //= 2` ‚Üí O(log n)
* `merge sort` / `quick sort` structure ‚Üí O(n log n)
