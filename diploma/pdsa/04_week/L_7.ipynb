**Topological Sorting** is a linear ordering of the **vertices** of a **Directed Acyclic Graph (DAG)** such that for every **directed edge** $u \rightarrow v$, vertex $u$ comes **before** vertex $v$ in the ordering.

---

### ðŸ”‘ **When Is Topological Sorting Possible?**

Topological sorting is only possible **if and only if** the graph is a **DAG** (Directed Acyclic Graph).
If the graph contains a **cycle**, topological sort **does not exist**.
