### üß† **Introduction to Directed Acyclic Graph (DAG)**

A **Directed Acyclic Graph (DAG)** is a **directed graph** that **has no cycles**.

---

### üîπ **Definition**

* A **graph** is a collection of **nodes (vertices)** connected by **edges**.
* A **directed graph** has edges with a direction (from one node to another).
* A **DAG** is a directed graph where:

  * Edges have a direction.
  * **No path** from a node leads **back to itself** (i.e., **no cycles**).

Formally, a graph $G = (V, E)$ is a **DAG** if there are **no directed cycles**.

---

### üîπ **Properties of DAGs**

1. ‚úÖ **No cycles** ‚Äî you cannot revisit the same node following the direction of edges.
2. ‚úÖ **Has at least one node with in-degree 0** (a starting point).
3. ‚úÖ **Topological sorting is possible** ‚Äî a linear ordering of nodes such that all edges go from earlier to later in the order.
4. ‚úÖ Used in many **real-world systems** where dependency ordering matters.

---

### üîπ **Real-World Examples of DAGs**

* **Task scheduling** (some tasks must be done before others).
* **Build systems** (e.g., CMake, Makefile dependencies).
* **Git version history** (commits form a DAG).
* **Course prerequisites** in curriculum design.
* **Expression trees** in compilers.

---

### üîπ **Topological Sort (Key Algorithm for DAG)**

A **topological sort** of a DAG is an ordering of vertices $v_1, v_2, ..., v_n$ such that **for every edge** $(u, v)$, $u$ comes **before** $v$.

This is used in:

* Resolving dependencies
* Pipeline execution
* Task sequencing

Time Complexity:

* Using DFS: $\mathcal{O}(V + E)$

---

### üîπ **How to Check if a Graph is a DAG?**

* Run **DFS** and check if there are **back edges**.
* If a back edge exists ‚Üí it‚Äôs **not** a DAG.
* If no back edge ‚Üí it's a DAG.
