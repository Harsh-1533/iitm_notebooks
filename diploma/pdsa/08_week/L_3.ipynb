## üî¢ **Problem Statement**

Given two large integers `x` and `y`, how can we multiply them faster than the traditional school method?

---

## üß† **1. The Naive (Grade-School) Multiplication**

In school, we learned to multiply numbers by forming partial products.

For example:

```
      23
    √ó 14
  _______
      92   ‚Üê 23 √ó 4 (unit digit)
+   230    ‚Üê 23 √ó 1 (ten digit, shifted)
  _______
     322
```

Each digit of the second number is multiplied with the entire first number, resulting in *n* partial products (if both have *n* digits), and then added together.

### üßÆ Time Complexity:

* If `x` and `y` are `n`-digit numbers (in base 10 or binary),
* Each partial product takes O(n),
* You generate and add `n` such products.

üî∫ **Total Time: O(n¬≤)**

---

## üí• **2. Binary Multiplication: Same Idea in Base 2**

Let‚Äôs multiply 12 √ó 13:

* 12 = `1100` (binary)
* 13 = `1101` (binary)

You form binary partial products:

```
        1100    ‚Üê 1 √ó 1100
+      0000     ‚Üê 0 √ó 1100
+     1100      ‚Üê 1 √ó 1100 (shifted 2 places)
+  1100         ‚Üê 1 √ó 1100 (shifted 3 places)
-------------
= 1111100 (binary) = 156 (decimal)
```

Again, you do `n` partial products of `n`-bit numbers, and add them ‚Üí **O(n¬≤)**.

---

## ‚úÇÔ∏è **3. Divide and Conquer Idea**

Just like Merge Sort, the core idea is:

* Split the problem into smaller subproblems
* Solve them recursively
* Combine results

So, for multiplying two `n`-bit numbers `x` and `y`, **split each number into two `n/2`-bit parts**:

```
x = x1 √ó 2^(n/2) + x0  
y = y1 √ó 2^(n/2) + y0
```

So,

```
x * y = (x1 √ó 2^m + x0)(y1 √ó 2^m + y0)
      = x1y1 √ó 2^(2m) + (x1y0 + x0y1) √ó 2^m + x0y0
```

### This requires:

* 4 multiplications of size `n/2`:

  * x1 \* y1
  * x0 \* y0
  * x1 \* y0
  * x0 \* y1
* And some additions and shifts (which take O(n))

### üßÆ Recurrence Relation:

```
T(n) = 4T(n/2) + O(n)
```

Using the **Master Theorem**:

* a = 4, b = 2, d = 1
* log\_b(a) = log‚ÇÇ4 = 2
* Since d < log\_b(a) ‚áí Case 1 applies

üî∫ **Time Complexity = O(n¬≤)** ‚Üí No improvement yet.

---

## üß†üí° **4. Karatsuba's Trick (1960)**

Karatsuba realized that **we don‚Äôt need all four multiplications**. We can **reduce it to three** using a clever algebraic trick.

Recall:

```
x * y = x1y1 √ó 2^2m + (x1y0 + x0y1) √ó 2^m + x0y0
```

Let:

* P = x1 \* y1
* Q = x0 \* y0
* R = (x1 + x0)(y1 + y0)

Then:

```
R = x1y1 + x1y0 + x0y1 + x0y0
So, R - P - Q = x1y0 + x0y1  ‚Üê cross terms!
```

Now plug in:

```
x * y = P √ó 2^2m + (R - P - Q) √ó 2^m + Q
```

### üßÆ We did:

* Only **3 multiplications**:

  * P = x1 \* y1
  * Q = x0 \* y0
  * R = (x1 + x0)(y1 + y0)
* Remaining is addition/subtraction and bit shifting ‚Üí O(n)

---

## üß†üìà **5. Karatsuba‚Äôs Recurrence Analysis**

```
T(n) = 3T(n/2) + O(n)
```

Use Master Theorem:

* a = 3, b = 2, d = 1
* log\_b(a) = log‚ÇÇ3 ‚âà 1.5849
* Since d < log\_b(a), again Case 1

üîª **Time Complexity = O(n^log‚ÇÇ3) ‚âà O(n^1.585)**

‚úÖ Much better than O(n¬≤) for large n!

---

## üß™ **6. Visualizing Recursive Steps**

For `T(n) = 3T(n/2) + n`, the recursion tree looks like this:

```
                T(n)
              /  |  \
           T(n/2)...(3 branches)
          / | \
      T(n/4)...(each has 3 branches)
        ...
     Total work = n at each level √ó log n levels
```

So,

* Total Work ‚âà O(n^log‚ÇÇ3)

---

## üïπÔ∏è **7. Algorithm Steps (Pseudocode)**

```python
def Karatsuba(x, y):
    # Base case for small numbers
    if x < 10 or y < 10:
        return x * y

    n = max(x.bit_length(), y.bit_length())
    m = n // 2

    # Split the numbers
    x1, x0 = x >> m, x & ((1 << m) - 1)
    y1, y0 = y >> m, y & ((1 << m) - 1)

    # 3 Recursive multiplications
    P = Karatsuba(x1, y1)
    Q = Karatsuba(x0, y0)
    R = Karatsuba(x1 + x0, y1 + y0)

    # Final product using Karatsuba‚Äôs formula
    return (P << (2 * m)) + ((R - P - Q) << m) + Q
```

## üìö **9. Historical Insight**

* 1960: Andrey Kolmogorov (famous Russian mathematician) conjectured that multiplication needed O(n¬≤) time.
* Karatsuba (his 23-year-old student) disproved it 2 weeks later!
* This was one of the first examples showing divide and conquer could beat na√Øve methods.

üîÅ Donald Knuth later simplified the analysis by using subtraction instead of addition to avoid bit overflows.

## üßÆ **10. Base Considerations**

The algorithm is base-independent:

* Works in **binary**, **decimal**, **hex**, etc.
* As long as you can:

  * Split numbers into halves
  * Multiply, add, and shift efficiently
