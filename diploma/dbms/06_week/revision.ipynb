Sure! Here’s a clear explanation of the **conditions for each normal form (1NF to 5NF and BCNF):**

---

## **1. First Normal Form (1NF)**

**Condition:**

* The domain of each attribute must contain only **atomic (indivisible) values**.
* There are **no repeating groups or arrays**.
* Each record must be unique (tuples are distinct).

**Example:**
❌ `PhoneNumbers = {12345, 67890}` (multi-valued)
✅ Split into separate rows or columns, each cell has a single value.

---

## **2. Second Normal Form (2NF)**

**Condition:**

* Must be in **1NF**, and
* **No partial dependency:** No non-prime attribute (non-key attribute) should depend on a **part of a composite candidate key**.

**Key point:**

* If the relation has a single attribute candidate key, 2NF is automatically satisfied.

**Example:**
❌ In `Student(CourseID, StudentID, CourseName)`: `CourseID → CourseName` depends only on part of composite key `(CourseID, StudentID)`.
✅ Move `CourseID → CourseName` to a separate table.

---

## **3. Third Normal Form (3NF)**

**Condition:**

* Must be in **2NF**, and
* **No transitive dependency:** A non-prime attribute must not depend on another non-prime attribute.
* Equivalently: For every FD `X → Y`, either:

  1. X is a superkey, or
  2. Each attribute in Y is a prime attribute (part of some candidate key).

**Example:**
❌ `StudentID → DeptID → DeptName` (DeptName depends on DeptID, which depends on StudentID).
✅ Remove transitive dependency by separating Department info.

---

## **Boyce–Codd Normal Form (BCNF)**

**Condition:**

* Stronger than 3NF.
* For every FD `X → Y`, **X must be a superkey**.

**Key difference from 3NF:**

* 3NF allows FDs where RHS is prime even if LHS is not a superkey.
* BCNF does **not** allow that.

**Example:**
❌ `Non-superkey → attribute`, even if RHS is prime.
✅ Decompose relation.

---

## **4. Fourth Normal Form (4NF)**

**Condition:**

* Must be in **BCNF**, and
* **No non-trivial multi-valued dependencies (MVDs)** except those that are implied by candidate keys.

**Example:**
If `Student →→ Course` and `Student →→ Hobby` (independent sets), this violates 4NF because MVDs lead to unnecessary combinations.

---

## **5. Fifth Normal Form (5NF) / Project-Join Normal Form (PJNF)**

**Condition:**

* Must be in **4NF**, and
* The relation cannot be decomposed into smaller relations **without losing data or introducing spurious tuples**.
* All possible join dependencies must be implied by candidate keys.

**Example:**
If a table can be losslessly split into three or more tables and then recombined using joins, you must do so.

---

### **Summary Chart**

| Normal Form | Condition                                              |
| ----------- | ------------------------------------------------------ |
| **1NF**     | Atomic values only, no repeating groups                |
| **2NF**     | 1NF + No partial dependency on part of a composite key |
| **3NF**     | 2NF + No transitive dependency (non-prime → non-prime) |
| **BCNF**    | For every FD X → Y, X must be a superkey               |
| **4NF**     | BCNF + No non-trivial multi-valued dependencies        |
| **5NF**     | 4NF + No non-trivial join dependencies                 |

Sure! Here are the **conditions for all normal forms (1NF to 5NF and BCNF) in a compact way**:

---

### **1NF (First Normal Form)**

* All attributes must have **atomic (indivisible) values**.
* No **repeating groups** or arrays.
* Each row must be **unique**.

---

### **2NF (Second Normal Form)**

* Must be in **1NF**, and
* **No partial dependency:**

  * Non-prime attributes must depend on the **whole candidate key**, not just part of it.

---

### **3NF (Third Normal Form)**

* Must be in **2NF**, and
* **No transitive dependency:**

  * Non-prime attributes must not depend on another non-prime attribute.
* Or: For every FD `X → Y`, either:

  1. **X is a superkey**, or
  2. **Y is a prime attribute** (part of some candidate key).

---

### **BCNF (Boyce–Codd Normal Form)**

* For every FD `X → Y`, **X must be a superkey**.
* Stronger than 3NF (doesn’t allow exceptions for prime attributes).

---

### **4NF (Fourth Normal Form)**

* Must be in **BCNF**, and
* **No non-trivial multi-valued dependencies (MVDs)** except those implied by candidate keys.

---

### **5NF (Fifth Normal Form / Project-Join Normal Form)**

* Must be in **4NF**, and
* **No non-trivial join dependencies** other than those implied by candidate keys.
* Relation cannot be decomposed into smaller relations without lossless join.
