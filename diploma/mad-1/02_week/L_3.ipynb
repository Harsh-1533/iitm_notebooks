# L2.3: What is Markup?

**Learning Outcome:**  
Understand how semantic meaning is added to raw content with the help of markup.

---

## üìÑ Content vs Meaning

- **Content** refers to the raw data or text (e.g., "Chapter 1", "Welcome to our site").
- **Meaning** is the *structure*, *purpose*, or *role* that content plays (e.g., "Chapter 1" is a heading, "Welcome to our site" is a paragraph).

üß† **Markup** adds **semantic meaning** to the content so computers and browsers can understand how to display or interpret it.

Example:
```html
<h1>Welcome to Our Site</h1>
<p>This is the home page of our website.</p>

- Here, <h1> adds meaning: it shows the text is a main heading, and <p> means it is a paragraph.
```

## üß∑ Types of Markup
1. Presentational Markup
- Focuses on how the content looks (e.g., font size, bold).
- Example: <b>Bold Text</b>, <i>Italic</i>
- Not commonly used anymore for structure.
2. Procedural Markup
- Tells a program how to process or format the document.
- Example: LaTeX commands in academic writing.
3. Descriptive (Semantic) Markup
- Describes the meaning of the content.
- Example: 
```html 
<header>, <article>, <section> in HTML5
```

## üåê HTML: The Most Common Markup Language
- HTML (HyperText Markup Language) is a semantic markup language used to structure web content.
- It uses tags to mark elements such as headings, paragraphs, links, lists, images, etc.
- Example:

```html
<!DOCTYPE html>
<html>
  <head>
    <title>My Page</title>
  </head>
  <body>
    <h1>Hello, World!</h1>
    <p>This is a simple HTML page.</p>
  </body>
</html>
```

## ‚úÖ Summary
- Markup separates content from meaning.
- It helps browsers and machines interpret content structure.
- HTML is the most widely used markup language for the web.
