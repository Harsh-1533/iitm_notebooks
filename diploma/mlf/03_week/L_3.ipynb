## âœ… 1. **Projection of a Vector onto a Line**

### ðŸ”· Goal:

Given a vector $\vec{b}$ and a line defined by a vector $\vec{a}$, we want to **project $\vec{b}$ onto the line in the direction of $\vec{a}$**.

### ðŸ§® Formula:

$$
\text{proj}_{\vec{a}} \vec{b} = \frac{\vec{a}^T \vec{b}}{\vec{a}^T \vec{a}} \cdot \vec{a}
$$

### ðŸ“Œ Example:

Let

$$
\vec{a} = \begin{bmatrix} 3 \\ 4 \end{bmatrix}, \quad
\vec{b} = \begin{bmatrix} 4 \\ 3 \end{bmatrix}
$$

**Step 1: Compute dot products**

$$
\vec{a}^T \vec{b} = 3 \cdot 4 + 4 \cdot 3 = 12 + 12 = 24
$$

$$
\vec{a}^T \vec{a} = 3^2 + 4^2 = 9 + 16 = 25
$$

**Step 2: Plug into formula**

$$
\text{proj}_{\vec{a}} \vec{b} = \frac{24}{25} \cdot \begin{bmatrix} 3 \\ 4 \end{bmatrix} = \begin{bmatrix} \frac{72}{25} \\ \frac{96}{25} \end{bmatrix}
$$

This is the projection of $\vec{b}$ **onto the line through** $\vec{a}$.

## âœ… 2. **Projection of a Vector onto Column Space of a Matrix**

Let matrix $A \in \mathbb{R}^{m \times n}$, and let $\vec{b} \in \mathbb{R}^m$.
You want to **project $\vec{b}$** onto the **column space of $A$**.

This is equivalent to solving the least squares problem:

$$
\min_{\vec{x}} \| A\vec{x} - \vec{b} \|
$$

### ðŸ§® Projection Formula:

$$
\text{proj}_{C(A)} \vec{b} = A (A^T A)^{-1} A^T \vec{b}
$$

---

### ðŸ“Œ Example:

Let

$$
A = \begin{bmatrix} 1 \\ 2 \end{bmatrix}, \quad
\vec{b} = \begin{bmatrix} 3 \\ 1 \end{bmatrix}
$$

**Step 1:** Compute $A^T A$

$$
A^T A = \begin{bmatrix} 1 & 2 \end{bmatrix} \begin{bmatrix} 1 \\ 2 \end{bmatrix} = [1^2 + 2^2] = [5]
$$

**Step 2:** Compute $A^T \vec{b}$

$$
A^T \vec{b} = \begin{bmatrix} 1 & 2 \end{bmatrix} \begin{bmatrix} 3 \\ 1 \end{bmatrix} = [3 + 2] = [5]
$$

**Step 3:** Compute projection

$$
\text{proj}_{C(A)} \vec{b} = A \cdot (A^T A)^{-1} A^T \vec{b}
= \begin{bmatrix} 1 \\ 2 \end{bmatrix} \cdot \left( \frac{1}{5} \cdot 5 \right)
= \begin{bmatrix} 1 \\ 2 \end{bmatrix}
$$

âœ… So, the projection of $\vec{b}$ onto column space of $A$ is $\begin{bmatrix} 1 \\ 2 \end{bmatrix}$

---

## âœ… 3. **Constructing the Projection Matrix**

The **projection matrix** $P$ that projects any $\vec{b}$ onto the **column space of $A$** is:

$$
P = A (A^T A)^{-1} A^T
$$

### ðŸ“Œ Example (continued):

$$
A = \begin{bmatrix} 1 \\ 2 \end{bmatrix}
\Rightarrow
P = \begin{bmatrix} 1 \\ 2 \end{bmatrix}
\cdot \left( \frac{1}{5} \right) \cdot \begin{bmatrix} 1 & 2 \end{bmatrix}
= \frac{1}{5} \begin{bmatrix} 1 \\ 2 \end{bmatrix} \begin{bmatrix} 1 & 2 \end{bmatrix}
= \frac{1}{5} \begin{bmatrix} 1 & 2 \\ 2 & 4 \end{bmatrix}
$$

So the projection matrix is:

$$
P = \begin{bmatrix} \frac{1}{5} & \frac{2}{5} \\ \frac{2}{5} & \frac{4}{5} \end{bmatrix}
$$

âœ… You can now compute the projection of **any vector $\vec{b}$** by just doing $P \vec{b}$

## ðŸ§© **Question:**

1. Find the **projection** of $\vec{b}$ onto the **column space** of $A$

3. Verify that $P \vec{b}$ gives the same result as part 1

## âœ… **Solution:**

### ðŸ”¹ Step 1: Projection of $\vec{b}$ onto $C(A)$

We use the formula:

$$
\text{proj}_{C(A)} \vec{b} = A (A^T A)^{-1} A^T \vec{b}
$$

### ðŸ§® Step-by-step:

#### 1. Compute $A^T A$:

$$
A^T = \begin{bmatrix} 1 & 1 & 0 \\ 0 & 1 & 1 \end{bmatrix}
\Rightarrow
A^T A = \begin{bmatrix}
1+1+0 & 0+1+0 \\
0+1+0 & 0+1+1
\end{bmatrix}
=
\begin{bmatrix}
2 & 1 \\
1 & 2
\end{bmatrix}
$$

#### 2. Compute $(A^T A)^{-1}$:

Use the inverse formula for 2Ã—2 matrix:

$$
(A^T A)^{-1} = \frac{1}{(2)(2)-(1)(1)} \begin{bmatrix} 2 & -1 \\ -1 & 2 \end{bmatrix}
= \frac{1}{3} \begin{bmatrix} 2 & -1 \\ -1 & 2 \end{bmatrix}
$$

#### 3. Compute $A^T \vec{b}$:

$$
A^T \vec{b} = \begin{bmatrix} 1 & 1 & 0 \\ 0 & 1 & 1 \end{bmatrix}
\begin{bmatrix} 2 \\ 3 \\ 4 \end{bmatrix}
= \begin{bmatrix} 2+3+0 \\ 0+3+4 \end{bmatrix}
= \begin{bmatrix} 5 \\ 7 \end{bmatrix}
$$

#### 4. Put it all together:

$$
\text{proj}_{C(A)} \vec{b} = A \cdot \left( \frac{1}{3} \begin{bmatrix} 2 & -1 \\ -1 & 2 \end{bmatrix} \right) \cdot \begin{bmatrix} 5 \\ 7 \end{bmatrix}
$$

First multiply:

$$
\frac{1}{3} \begin{bmatrix} 2 & -1 \\ -1 & 2 \end{bmatrix} \begin{bmatrix} 5 \\ 7 \end{bmatrix}
= \frac{1}{3} \begin{bmatrix} 2 \cdot 5 - 1 \cdot 7 \\ -1 \cdot 5 + 2 \cdot 7 \end{bmatrix}
= \frac{1}{3} \begin{bmatrix} 10 - 7 \\ -5 + 14 \end{bmatrix}
= \frac{1}{3} \begin{bmatrix} 3 \\ 9 \end{bmatrix}
= \begin{bmatrix} 1 \\ 3 \end{bmatrix}
$$

Now multiply with $A$:

$$
A \cdot \begin{bmatrix} 1 \\ 3 \end{bmatrix}
= \begin{bmatrix} 1 \cdot 1 + 0 \cdot 3 \\ 1 \cdot 1 + 1 \cdot 3 \\ 0 \cdot 1 + 1 \cdot 3 \end{bmatrix}
= \begin{bmatrix} 1 \\ 4 \\ 3 \end{bmatrix}
$$

âœ… Final projection result:

$$
\text{proj}_{C(A)} \vec{b} = \begin{bmatrix} 1 \\ 4 \\ 3 \end{bmatrix}
$$

### ðŸ”¹ Step 2: Projection Matrix $P$

$$
P = A (A^T A)^{-1} A^T = A \cdot \left( \frac{1}{3} \begin{bmatrix} 2 & -1 \\ -1 & 2 \end{bmatrix} \right) \cdot A^T
$$

We already know:

$$
A = \begin{bmatrix} 1 & 0 \\ 1 & 1 \\ 0 & 1 \end{bmatrix}, \quad
A^T = \begin{bmatrix} 1 & 1 & 0 \\ 0 & 1 & 1 \end{bmatrix}

So you can multiply the full product to get a 3Ã—3 matrix $P$. You can also use software (NumPy, MATLAB, etc.) to compute it if needed.

If you compute the matrix $P$ and apply it to $\vec{b}$, you'll get the same result:

$$
P \vec{b} = \begin{bmatrix} 1 \\ 4 \\ 3 \end{bmatrix}
$$
