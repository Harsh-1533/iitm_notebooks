# **Title: Understanding the Four Fundamental Subspaces of a Matrix (with Examples)**

---

## üìò Introduction

In linear algebra, every matrix gives rise to **four fundamental subspaces** that help us understand its structure:

1. **Column Space (C(A))**
2. **Null Space (N(A))**
3. **Row Space (C(A·µó))**
4. **Left Null Space (N(A·µó))**

These spaces relate to each other through the **Rank-Nullity Theorem** and reveal important geometric and algebraic properties of the matrix.

---

## ‚úâÔ∏è Given Matrix Example

Let:

```markdown
A = 
[  1   3   3   2  
   2   6   9   7  
  -1  -3   3   4 ]
```

This is a 3x4 matrix.

---

## üìä Step 1: Row Reduce A

We perform Gaussian elimination to bring A into row echelon form (REF):

After row operations, we get:

```markdown
REF of A =
[ 1   3   3   2 
  0   0   3   3 
  0   0   0   0 ]
```

* **Pivot Columns**: Column 1, Column 3
* **Rank(A) = 2**

---

## 1. Column Space C(A)

The column space of A is the span of the **pivot columns** of the original matrix A:

```markdown
C(A) = span{ [1 2 -1]·µó, [3 9 3]·µó }
```

It is a **2D subspace of** ‚Ñù¬≥.

**üîÅ Remark:**

> The column space **C(A)** is **orthogonal** to the **left null space** **N(A·µó)**.

---

## 2. Null Space N(A)

Solve A¬∑x = 0. From REF, the system becomes:

```markdown
x‚ÇÅ + 3x‚ÇÇ + 3x‚ÇÉ + 2x‚ÇÑ = 0
3x‚ÇÉ + 3x‚ÇÑ = 0  ‚Üí  x‚ÇÉ = -x‚ÇÑ
```

Substitute:

```markdown
x‚ÇÅ = -3x‚ÇÇ + x‚ÇÑ
```

Let:

* x‚ÇÇ = s (free variable)
* x‚ÇÑ = t (free variable)

Then general solution:

```markdown
x = s * [-3 1 0 0]·µó + t * [1 0 -1 1]·µó
```

```markdown
N(A) = span{ [-3 1 0 0]·µó, [1 0 -1 1]·µó }
```

This is a **2D subspace of** ‚Ñù‚Å¥.

---

## 3. Row Space C(A·µó)

The row space is the span of the **nonzero rows** of REF of A:

```markdown
C(A·µó) = span{ [1 3 3 2], [0 0 1 1] }
```

This is a **2D subspace of** ‚Ñù‚Å¥.

**üîÅ Remark:**

> The row space **C(A·µó)** is **orthogonal** to the **null space** **N(A)**.

---

## 4. Left Null Space N(A·µó)

This is the set of all vectors y such that:

```markdown
A·µó¬∑y = 0
```

Since A is 3x4, A·µó is 4x3. Use rank-nullity:

* Rank of A·µó = Rank of A = 2
* Nullity(A·µó) = 3 - 2 = 1

So the left null space is a **1D subspace of** ‚Ñù¬≥.
