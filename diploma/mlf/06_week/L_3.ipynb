## ðŸ“˜ **Topic: Positive Definiteness**

### ðŸ§­ **1. Motivation: Quadratic Functions and Stationary Points**

We start with a **quadratic function** in two variables:

$$
f(x, y) = 2x^2 + 4xy + y^2
$$

#### ðŸ‘‰ **Stationary Points**

To find stationary points, compute the **first partial derivatives**:

* $\frac{\partial f}{\partial x} = 4x + 4y$
* $\frac{\partial f}{\partial y} = 4x + 2y$

At the **stationary point**, both derivatives vanish:

$$
\frac{\partial f}{\partial x} = \frac{\partial f}{\partial y} = 0 \Rightarrow x = 0, y = 0
$$

So, $(0, 0)$ is a stationary point.

#### ðŸ§  **But is it a Minima, Maxima, or Saddle Point?**

To determine this, use the **second derivative test**:

* $\frac{\partial^2 f}{\partial x^2} = 4$
* $\frac{\partial^2 f}{\partial x \partial y} = 4$
* $\frac{\partial^2 f}{\partial y^2} = 2$

Form the **Hessian matrix**:

$$
H = \begin{bmatrix}
4 & 4 \\
4 & 2
\end{bmatrix}
$$

The nature of the stationary point depends on the definiteness of the **quadratic form** defined by this Hessian.

---

### ðŸ” **2. Definition: Positive Definite Function**

A function $f(x, y)$ is **positive definite** if:

* $f(0,0) = 0$
* $f(x, y) > 0$ for all $(x, y) \neq (0,0)$

ðŸ“Œ Example: $f(x, y) = x^2 + y^2$ is positive definite since it reaches minimum 0 at the origin and is strictly positive elsewhere.

---

### ðŸ§® **3. General Form of a Quadratic Function**

$$
f(x, y) = ax^2 + 2bxy + cy^2
$$

#### âœ… **Necessary Conditions for Positive Definiteness**

* $a > 0$: Since $f(1, 0) = a$, and we require $f > 0$ for all non-zero inputs.
* $c > 0$: Since $f(0, 1) = c$, and similarly it must be positive.

#### âŒ **But Not Sufficient**

Example:

$$
f(x, y) = x^2 - 10xy + y^2, \text{ where } a = 1, b = -5, c = 1
$$

Here:

* $a > 0$, $c > 0$
* But $f(1, 1) = 1 - 10 + 1 = -8 < 0$, so **not** positive definite

---

### ðŸ”Ž **4. Completing the Square (Algebraic Insight)**

Rewrite:

$$
f(x, y) = ax^2 + 2bxy + cy^2 = a\left(x + \frac{b}{a}y\right)^2 + \left(c - \frac{b^2}{a}\right)y^2
$$

For $f(x, y) > 0$ âˆ€ $x, y \neq 0$, both terms must be **strictly positive**:

* $a > 0$
* $c - \frac{b^2}{a} > 0 \Rightarrow ac > b^2$

---

### âœ… **5. Final Criteria for Positive Definiteness (2D Case)**

The quadratic function:

$$
f(x, y) = ax^2 + 2bxy + cy^2
$$

is **positive definite â‡”**

* $a > 0$
* $ac > b^2$

If:

* $ac = b^2$ â†’ **positive semi-definite**
* $ac < b^2$ â†’ **saddle point** at the origin

### ðŸ§® **6. Matrix Representation and Linear Algebra Connection**

Write:

$$
f(x, y) = \begin{bmatrix}x & y\end{bmatrix}
\begin{bmatrix}
a & b \\
b & c
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
= \vec{v}^\top A \vec{v}
$$

Where:

* $A = \begin{bmatrix} a & b \\ b & c \end{bmatrix}$
* $\vec{v} = \begin{bmatrix} x \\ y \end{bmatrix}$

This form generalizes to **n dimensions**:

$$
f(\vec{v}) = \vec{v}^\top A \vec{v}, \quad \text{where } \vec{v} \in \mathbb{R}^n
$$

---

### ðŸ”— **7. Definition: Positive Definite Matrix (Linear Algebra View)**

A **real symmetric matrix $A$** is **positive definite** if:

$$
\vec{v}^\top A \vec{v} > 0 \quad \text{for all } \vec{v} \neq \vec{0}
$$

* If $\vec{v}^\top A \vec{v} \geq 0$ â†’ **Positive Semi-Definite**
* If $\vec{v}^\top A \vec{v} < 0$ for all $\vec{v} \neq \vec{0}$ â†’ **Negative Definite**

### ðŸ§  **8. Summary of Important Points**

| Concept                        | Condition                                                                    |
| ------------------------------ | ---------------------------------------------------------------------------- |
| **Stationary point**           | First partial derivatives = 0                                                |
| **Minima**                     | Second derivative (Hessian) is positive definite                             |
| **Positive definite function** | $f(x, y) > 0$ for all $(x, y) \ne (0,0)$, and $f(0,0) = 0$                   |
| **Matrix positive definite**   | $A$ symmetric and $\vec{v}^\top A \vec{v} > 0$ for all $\vec{v} \ne \vec{0}$ |
| **2D case**                    | $a > 0$, $ac > b^2$                                                          |
| **Saddle point**               | Occurs when $ac < b^2$                                                       |
