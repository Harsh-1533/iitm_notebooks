## 📘 **Positive Definite Matrices – Detailed Notes**

### 🔰 **Context & Motivation**

* The concept of **positive definiteness** arises naturally in **optimization**, **machine learning**, and **quadratic forms**.
* Positive definiteness of functions has been previously introduced. This lecture shifts focus to **positive definiteness of real symmetric matrices**, a foundational concept in linear algebra.

---

## 🧩 **Definition: Positive Definite Matrix**

Let $A \in \mathbb{R}^{n \times n}$ be a **real symmetric matrix**.

### ✅ **A is called positive definite if:**

$$
\mathbf{v}^T A \mathbf{v} > 0 \quad \text{for all } \mathbf{v} \in \mathbb{R}^n, \mathbf{v} \ne \mathbf{0}
$$

This is called **Condition 1**.

---

## 📐 **Quadratic Form and Positive Definiteness**

Given a quadratic function:

$$
f(x, y) = ax^2 + 2bxy + cy^2
$$

The matrix representation of this function is:

$$
A = \begin{bmatrix}
a & b \\
b & c
\end{bmatrix}
$$

This function is positive definite **iff**:

* $a > 0$
* $ac - b^2 > 0$

These conditions ensure that:

* The matrix is symmetric.
* The corresponding quadratic form is strictly positive for all non-zero vectors.

---

## 📊 **Eigenvalue Interpretation**

Let $A$ be a real symmetric matrix with eigenvalues $\lambda_1, \lambda_2, \dots, \lambda_n$.

### ✅ **Condition 2: A is positive definite iff all eigenvalues are strictly positive:**

$$
\lambda_i > 0 \quad \text{for all } i = 1, 2, \dots, n
$$

---

## 🔄 **Equivalence Between the Two Definitions**

We aim to show:

* **Condition 1** $\iff$ **Condition 2**

---

### 🧠 **Proof: Condition 1 ⇒ Condition 2**

1. Assume $\mathbf{v}^T A \mathbf{v} > 0$ for all $\mathbf{v} \ne 0$.
2. Consider the eigenvalue equation: $A\mathbf{x} = \lambda \mathbf{x}$, for $\mathbf{x} \ne 0$.
3. Then:

   $$
   \mathbf{x}^T A \mathbf{x} = \lambda \mathbf{x}^T \mathbf{x} = \lambda \|\mathbf{x}\|^2
   $$
4. Since $\|\mathbf{x}\|^2 > 0$, and $\mathbf{x}^T A \mathbf{x} > 0$, it must be that $\lambda > 0$.
5. So, all eigenvalues are positive.

✅ **Condition 2 holds.**

---

### 🧠 **Proof: Condition 2 ⇒ Condition 1**

1. Assume $\lambda_i > 0$ for all eigenvalues of $A$.
2. Since $A$ is real symmetric, by the **Spectral Theorem**:

   * $A$ has an orthonormal basis of eigenvectors $\{\mathbf{x}_1, \dots, \mathbf{x}_n\}$.
3. Any vector $\mathbf{v} \in \mathbb{R}^n$ can be expressed as:

   $$
   \mathbf{v} = c_1 \mathbf{x}_1 + \dots + c_n \mathbf{x}_n
   $$
4. Then:

   $$
   \mathbf{v}^T A \mathbf{v} = \sum_{i=1}^{n} c_i^2 \lambda_i
   $$
5. Since each $\lambda_i > 0$, and $\mathbf{v} \ne 0 \Rightarrow \exists c_i \ne 0$, it follows that:

   $$
   \mathbf{v}^T A \mathbf{v} > 0
   $$

✅ **Condition 1 holds.**

---

## 🔍 **2×2 Matrix Case (Example)**

Given a matrix:

$$
A = \begin{bmatrix}
a & b \\
b & c
\end{bmatrix}
$$

This is **positive definite iff**:

1. $a > 0$
2. $ac - b^2 > 0$

These imply:

* $\text{det}(A) > 0$ → $\lambda_1 \cdot \lambda_2 > 0$
* $\text{trace}(A) = a + c > 0$ → $\lambda_1 + \lambda_2 > 0$

So both eigenvalues $\lambda_1, \lambda_2 > 0$

---

## 🧠 **Key Theoretical Takeaways**

* The **spectral theorem** is crucial in proving the equivalence.
* **Two equivalent definitions** of a positive definite real symmetric matrix:

  * $\mathbf{v}^T A \mathbf{v} > 0$ for all $\mathbf{v} \ne 0$
  * All eigenvalues of $A$ are strictly positive
