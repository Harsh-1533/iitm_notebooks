## ðŸ”¹ 1. Linear Approximation

A **linear approximation** estimates the value of a function near a point using the tangent line.

If a function \( f \) is differentiable at a point \( a \), then the linear approximation near \( x = a \) is:

$$
f(x) \approx f(a) + f'(a)(x - a)
$$

âœ… This is just the **tangent line** to the function at \( x = a \).

---

### ðŸ” Example:

Let  
$$
f(x) = \sqrt{1 + x}
$$

Approximate near \( x = 0 \):

- \( f(0) = \sqrt{1} = 1 \)
- \( f'(x) = \frac{1}{2\sqrt{1 + x}} \Rightarrow f'(0) = \frac{1}{2} \)

So,  
$$
f(x) \approx 1 + \frac{1}{2}x
$$

## ðŸ”¹ 2. Quadratic Approximation

To improve accuracy, we use a **second-degree Taylor polynomial**:

$$
f(x) \approx f(a) + f'(a)(x - a) + \frac{f''(a)}{2}(x - a)^2
$$

This includes **curvature** information.

---

### ðŸ” Example:

Let  
$$
f(x) = \sin(x)
$$

Approximate near \( x = 0 \):

- \( f(0) = 0 \)
- \( f'(x) = \cos(x) \Rightarrow f'(0) = 1 \)
- \( f''(x) = -\sin(x) \Rightarrow f''(0) = 0 \)

So,  
$$
f(x) \approx x
$$

## ðŸ”¹ 3. Chain Rule

If  
$$
y = f(g(x)),
$$

Then:  
$$
\frac{dy}{dx} = f'(g(x)) \cdot g'(x)
$$

---

### ðŸ” Example 1:

Differentiate:  
$$
f(x) = \frac{e^{3x}}{\sqrt{1 + x}} \quad \text{around } x = 0
$$

Let:
- \( u(x) = e^{3x} \)
- \( v(x) = \sqrt{1 + x} = (1 + x)^{1/2} \)

Use the quotient rule:
$$
f'(x) = \frac{u'(x)v(x) - u(x)v'(x)}{v(x)^2}
$$

Compute:
- \( u'(x) = 3e^{3x} \)
- \( v'(x) = \frac{1}{2}(1 + x)^{-1/2} \)

So:
$$
f'(x) = \frac{3e^{3x}\sqrt{1 + x} - e^{3x} \cdot \frac{1}{2\sqrt{1 + x}}}{1 + x}
$$

At \( x = 0 \):
$$
f'(0) = \frac{3e^0 \cdot 1 - e^0 \cdot \frac{1}{2}}{1} = \frac{3 - \frac{1}{2}}{1} = \frac{5}{2}
$$

âœ… Final answer:
$$
f'(0) = \frac{5}{2}
$$

---

### ðŸ” Example 2:

Differentiate:  
$$
f(x) = e^{\sqrt{1 + x}} \quad \text{around } x = 1
$$

Let:
- Inner: \( u(x) = \sqrt{1 + x} = (1 + x)^{1/2} \)
- Outer: \( f(u) = e^u \)

Then:
$$
f'(x) = e^{\sqrt{1 + x}} \cdot \frac{1}{2\sqrt{1 + x}}
$$

At \( x = 1 \), \( \sqrt{1 + x} = \sqrt{2} \):

$$
f'(1) = \frac{e^{\sqrt{2}}}{2\sqrt{2}}
$$

## ðŸ”¹ 4. Maxima, Minima, and Saddle Points

### âœ… First Derivative Test

At **critical points** where \( f'(x) = 0 \), use the second derivative:

- If \( f''(x) > 0 \): **Local Minimum**
- If \( f''(x) < 0 \): **Local Maximum**
- If \( f''(x) = 0 \): **Inconclusive** â†’ Saddle or Inflection Point

---

### ðŸ” Example:

Let  
$$
f(x) = x^3
$$

- \( f'(x) = 3x^2 \Rightarrow f'(0) = 0 \)
- \( f''(x) = 6x \Rightarrow f''(0) = 0 \)

So,  
$$
x = 0 \text{ is a saddle point (inflection point)}
$$
