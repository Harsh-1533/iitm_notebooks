# Geometry of Lines

## 1. A Line in $\mathbb{R}^a \subseteq \mathbb{R}^d$

A line in $\mathbb{R}^a$ is also a subset of $\mathbb{R}^d$. That is, any lower-dimensional line is embedded in higher-dimensional space.

---

## 2. Line Through a Point Along a Vector

Let:

* $u \in \mathbb{R}^d$: a point
* $v \in \mathbb{R}^d$: a direction vector

Then, the line passing through point $u$ in the direction of vector $v$ is defined as:

$$
\{ x \in \mathbb{R}^d : x = u + \alpha v, \ \alpha \in \mathbb{R} \}
$$

This describes all points on the line obtained by moving from $u$ in the direction of $v$, scaled by $\alpha$.

---

## 3. Line Through Two Points

Let:

* $u \in \mathbb{R}^d$
* $u' \in \mathbb{R}^d$

Then the direction vector is:

$$
v = u' - u
$$

### First Form

The line through $u$ and $u'$ is given by:

$$
\{ x \in \mathbb{R}^d : x = u + \alpha (u' - u), \ \alpha \in \mathbb{R} \}
$$

This form starts at $u$ and moves in the direction toward $u'$.

### Second (Interpolated) Form

The same line can also be written as:

$$
\{ x \in \mathbb{R}^d : x = (1 - \alpha)u + \alpha u', \ \alpha \in \mathbb{R} \}
$$

This is a linear interpolation:

* $\alpha = 0 \Rightarrow x = u$
* $\alpha = 1 \Rightarrow x = u'$
* $0 < \alpha < 1 \Rightarrow x$ is between $u$ and $u'$
* $\alpha < 0$ or $\alpha > 1 \Rightarrow x$ is beyond the segment

---

## 4. Directional Notes

* "Line through $u$ along $u' - u$" means: start at $u$, move toward $u'$
* "Line through $u'$ along $u - u'$" means: start at $u'$, move in reverse toward $u$

Both define the same line but from different reference points.

# Geometry of (Hyper)planes

A $(d-1)$-dimensional **hyperplane** $\subseteq \mathbb{R}^d$

A hyperplane **normal** to the vector $w \in \mathbb{R}^d$ with value $b \in \mathbb{R}$ is defined as:

$$
\{ x \in \mathbb{R}^d : w^T x = b \}
$$

Or written as a summation:

$$
\left\{ x \in \mathbb{R}^d : \sum_{i=1}^{d} w_i x_i = b \right\}
$$

This represents the set of points $x \in \mathbb{R}^d$ that lie on the hyperplane orthogonal to vector $w$ at offset $b$.


# Partial Derivatives and Gradients

## 1. Partial Derivatives

Partial derivatives describe how a multivariable function changes when **only one variable** changes, keeping the others constant.

Let $f(x, y)$ be a function of two variables. The partial derivatives of $f$ are:

* $\frac{\partial f}{\partial x}$: Rate of change of $f$ with respect to $x$, keeping $y$ constant.
* $\frac{\partial f}{\partial y}$: Rate of change of $f$ with respect to $y$, keeping $x$ constant.

### Example 1:

Let $f(x, y) = x^2y + 3xy^2$

* $\frac{\partial f}{\partial x} = 2xy + 3y^2$
* $\frac{\partial f}{\partial y} = x^2 + 6xy$

---

## 2. Gradient

The **gradient** of a scalar-valued function $f(x_1, x_2, \dots, x_d)$ is a **vector** of all its partial derivatives:

$$
\nabla f(x) = \left[ \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \dots, \frac{\partial f}{\partial x_d} \right]^T
$$

It points in the direction of **steepest ascent** of the function.

### Example 2:

Let $f(x, y) = x^2 + y^2$

* $\nabla f(x, y) = \left[ \frac{\partial f}{\partial x}, \frac{\partial f}{\partial y} \right] = [2x, 2y]$

### Example 3:

Let $f(x, y, z) = xyz$

* $\frac{\partial f}{\partial x} = yz$
* $\frac{\partial f}{\partial y} = xz$
* $\frac{\partial f}{\partial z} = xy$

So,

$$
\nabla f(x, y, z) = [yz, xz, xy]
$$
