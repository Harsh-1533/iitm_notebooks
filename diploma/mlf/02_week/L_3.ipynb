## üîπ 1. What is a Derivative?

A **derivative** tells us **how fast a function is changing** at a specific point.

Think of it like the **slope of the curve** (instantaneous rate of change).

### üéØ Formal definition:
If \( f(x) \) is a function, then the derivative at \( x = a \) is:

$$
f'(a) = \lim_{h \to 0} \frac{f(a+h) - f(a)}{h}
$$

---

## üîç Example:

Let $$ f(x) = x^2 $$

We want to find the derivative at \( x = 2 \).

We use:

$$
f'(x) = \frac{d}{dx}(x^2) = 2x
$$

So,

$$
f'(2) = 2 \cdot 2 = 4
$$

‚úÖ This means: *at \( x = 2 \), the function is increasing with slope 4.*

## üîπ 2. What is Linear Approximation?

A **linear approximation** is using the **tangent line** at a point to estimate values of the function nearby.

Think of it as saying:
> "The function is hard to calculate, but the **line is easier**. Let's use that."

---

### üìò Formula for Linear Approximation:

If $ f(x) $ is differentiable at \( x = a \), then near \( x = a \):

$$
f(x) \approx f(a) + f'(a)(x - a)
$$

This is just the **equation of the tangent line** at \( x = a \).

---

## üîç Example:

Let \( f(x) = \sqrt{x} \), and we want to estimate \( \sqrt{4.1} \)

Let‚Äôs use \( a = 4 \), because we know \( \sqrt{4} = 2 \)

Now:

- \( f(x) = \sqrt{x} \)
- \( f'(x) = \frac{1}{2\sqrt{x}} \)
- \( f'(4) = \frac{1}{2 \cdot 2} = \frac{1}{4} \)

So,

$$
f(x) \approx f(4) + f'(4)(x - 4)
$$

$$
f(4.1) \approx 2 + \frac{1}{4}(4.1 - 4) = 2 + \frac{1}{4}(0.1) = 2 + 0.025 = 2.025
$$

‚úÖ Actual \( \sqrt{4.1} \approx 2.0248 \), so linear approximation is **very close!**

## üí° When to Use Linear Approximation?

- When function values are hard to compute exactly
- When you're estimating near a known point
- In real-life physics, engineering, and machine learning problems
