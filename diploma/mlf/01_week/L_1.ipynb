- Machine Learning (ML) is the study of computer algorithms that improve automatically through experience and by th use of data.

#### This diagram shows a task hierarchy comparing how work is done in Machine Learning, Programming, and Manual Labour:

1. Machine Learning (Top Layer)
    - Human designs the tool (the ML model) using data.
    - Once trained, the tool (model) takes input and produces output automatically. 
    - Key point: Human effort goes into designing the system, not running it.

2. Programming & Tool Usage (Middle Layer)
    - A human uses a tool (like software) to process input and produce output.  
    - The tool doesnt learn; it just follows fixed instructions written by the human.
    - Key point: The human is actively involved each time.

3. Manual Labour (Bottom Layer)
    - The human is the tool: input is given directly to a human who creates the output.
    - No automation or external tools.
    - Key point: Fully human-driven task.

### Summary:
   - ML = Tool learns from data; minimal human effort once trained.
   - Programming = Human uses fixed tools; needs human input each time.
   - Manual Labour = Human does all the work directly.

### Why and When Machine Learning?
   - 1. Programming/Human Labour Fails
        1. Scale/Speed/Cost of human labor
        2. Inability to express rules using language.
        3. Don't know the exact rules transforming input to output.
   - 2. Machine learning can succeed.
        1. Have lots of example data
        2. Have some structural idea on the rules
