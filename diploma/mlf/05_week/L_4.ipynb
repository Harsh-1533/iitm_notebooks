## ‚úÖ **Topic: Diagonalization of Hermitian Matrices**

---

### üî∑ **Key Definitions**

1. **Hermitian Matrix (A):**

   * A complex square matrix is **Hermitian** if:

     $$
     A^* = A
     $$

     where $A^*$ is the **conjugate transpose** of $A$.

2. **Unitary Matrix (U):**

   * A complex square matrix is **unitary** if:

     $$
     U^*U = UU^* = I
     $$

     * Its inverse equals its conjugate transpose: $U^{-1} = U^*$

3. **Unitarily Diagonalizable Matrix:**

   * A matrix $A$ is **unitarily diagonalizable** if:

     $$
     A = U \Lambda U^*
     $$

     where:

     * $U$ is a unitary matrix,
     * $\Lambda$ is a diagonal matrix of eigenvalues of $A$.

---

### üî∑ **Goal of the Lecture**

* To prove that:

  $$
  \text{Any Hermitian matrix } A \text{ is unitarily diagonalizable.}
  $$

  * This will be done by using **Schur‚Äôs Theorem** and leveraging properties of Hermitian matrices.

---

## ‚úÖ **Schur‚Äôs Theorem**

### üìå **Statement:**

* **Any** square complex matrix $A \in \mathbb{C}^{n \times n}$ can be written as:

  $$
  A = U T U^*
  $$

  where:

  * $U$ is a **unitary matrix**
  * $T$ is an **upper triangular matrix**
  * $A$ and $T$ are **unitarily similar**

---

### üß† **Why It Matters:**

* Schur‚Äôs Theorem helps **reduce any matrix** to upper triangular form using unitary matrices.
* We can then use this result to prove **Hermitian matrices** are **diagonalizable**.

---

### üß© **Key Idea Behind the Proof (For $n = 3$):**

1. **Start with Eigenvalue Œª‚ÇÅ** and its eigenvector $z_1$.

2. **Extend** $z_1$ to an orthonormal basis using **Gram-Schmidt**.

3. Form a **unitary matrix $U_1$** with these basis vectors as columns.

4. Compute:

   $$
   U_1^* A U_1 = \begin{bmatrix}
   \lambda_1 & * & * \\
   0 & * & * \\
   0 & 0 & *
   \end{bmatrix}
   $$

   * This gives a **partially upper triangular form.**

5. **Repeat the process** recursively on the lower block until the whole matrix is upper triangular.

6. Final unitary matrix:

   $$
   U = U_1 U_2 \dots U_k
   \Rightarrow A = U T U^*
   $$

---

## ‚úÖ **Special Case: Hermitian Matrices**

* If $A = A^*$, i.e., **Hermitian**, then:

  * The upper triangular matrix $T$ must be **diagonal**.
  * Hence, Hermitian matrices are **unitarily diagonalizable**.

---

### ‚úÖ **Final Result (Spectral Theorem):**

For any **Hermitian matrix** $A$, there exists a **unitary matrix** $U$ such that:

$$
A = U \Lambda U^*
$$

* $\Lambda$ is a **real diagonal matrix** (eigenvalues are real)
* Columns of $U$ are the **orthonormal eigenvectors** of $A$

---

### üîç **Example Outline (3√ó3 Matrix):**

Given matrix:

$$
A = \begin{bmatrix}
5 & 8 & 16 \\
5 & 0 & 9 \\
-3 & -5 & -10
\end{bmatrix}
$$

Steps:

1. Find eigenvalues using the **characteristic polynomial**.
2. Find an **eigenvector** for one eigenvalue (say, $\lambda_1 = 1$).
3. Extend it to an **orthonormal basis** using Gram-Schmidt.
4. Form matrix $U_1$ and compute $U_1^*AU_1$.
5. Focus on the smaller bottom-right block and repeat with it.
6. Combine all unitary matrices into $U = U_1 U_2$, and compute final:

   $$
   U^* A U = T \text{ (Upper Triangular)}
   $$

## ‚úÖ **Example: Diagonalizing a Hermitian Matrix**

Let‚Äôs consider the matrix:

$$
A = \begin{bmatrix}
2 & i \\
-i & 2
\end{bmatrix}
$$

### Step 1: **Check if the matrix is Hermitian**

A matrix $A$ is Hermitian if $A^* = A$, where $A^*$ is the **conjugate transpose**.

* $A^* = \begin{bmatrix}
  2 & -i \\
  i & 2
  \end{bmatrix}$

* And since $A^* = A$, ‚úÖ **Yes, it‚Äôs Hermitian**

---

### Step 2: **Find the eigenvalues**

We solve the characteristic equation:

$$
\det(A - \lambda I) = 0
$$

$$
A - \lambda I = \begin{bmatrix}
2 - \lambda & i \\
-i & 2 - \lambda
\end{bmatrix}
$$

$$
\text{Det} = (2 - \lambda)^2 - (i)(-i) = (2 - \lambda)^2 - 1 = 0
$$

$$
(2 - \lambda)^2 = 1 \Rightarrow 2 - \lambda = \pm 1 \Rightarrow \lambda = 1, 3
$$

‚úÖ Eigenvalues: $\lambda_1 = 1$, $\lambda_2 = 3$

---

### Step 3: **Find the eigenvectors**

#### For $\lambda = 1$:

Solve:

$$
(A - I)v = 0 \Rightarrow \begin{bmatrix}
1 & i \\
-i & 1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
= 0
$$

Solving this, we get:

$$
x + i y = 0 \Rightarrow x = -iy
$$

Let‚Äôs choose $y = 1 \Rightarrow x = -i$

Eigenvector:

$$
v_1 = \begin{bmatrix}
-i \\
1
\end{bmatrix}
$$

Let‚Äôs normalize it:

$$
\|v_1\| = \sqrt{(-i)^*\cdot(-i) + (1)^*\cdot(1)} = \sqrt{1 + 1} = \sqrt{2}
$$

So:

$$
u_1 = \frac{1}{\sqrt{2}} \begin{bmatrix}
-i \\
1
\end{bmatrix}
$$

---

#### For $\lambda = 3$:

Solve:

$$
(A - 3I)v = 0 \Rightarrow \begin{bmatrix}
-1 & i \\
-i & -1
\end{bmatrix}
\begin{bmatrix}
x \\
y
\end{bmatrix}
= 0
$$

$$
-x + i y = 0 \Rightarrow x = i y
\Rightarrow y = 1 \Rightarrow x = i
$$

Eigenvector:

$$
v_2 = \begin{bmatrix}
i \\
1
\end{bmatrix}
$$

Normalize it:

$$
\|v_2\| = \sqrt{i^*i + 1^*1} = \sqrt{1 + 1} = \sqrt{2}
\Rightarrow u_2 = \frac{1}{\sqrt{2}} \begin{bmatrix}
i \\
1
\end{bmatrix}
$$

---

### Step 4: **Form the unitary matrix U and diagonal matrix Œõ**

Let:

$$
U = [u_1 \quad u_2] = \frac{1}{\sqrt{2}} \begin{bmatrix}
-i & i \\
1 & 1
\end{bmatrix}
$$

$$
\Lambda = \begin{bmatrix}
1 & 0 \\
0 & 3
\end{bmatrix}
$$

‚úÖ Final decomposition:

$$
A = U \Lambda U^*
$$

---

### üîç Final Check:

Let‚Äôs multiply $U \Lambda U^*$ and verify that we get back $A$. You can do this manually or in Python/NumPy.

## ‚úÖ Summary

* Matrix $A = \begin{bmatrix} 2 & i \\ -i & 2 \end{bmatrix}$ is Hermitian.
* Eigenvalues: $\lambda = 1, 3$
* Eigenvectors were orthonormalized to form unitary matrix $U$.
* Final form: $A = U \Lambda U^*$

## üìå **Spectral Theorem Summary**

### ‚úÖ **Key Result:**

> A **Hermitian matrix** $A$ is **unitarily diagonalizable**.

This means:

$$
A = U \Lambda U^*
$$

* $U$: a **unitary matrix** (columns are orthonormal eigenvectors)
* $\Lambda$: a **diagonal matrix** (real eigenvalues of $A$)
* $U^*$: conjugate transpose of $U$

---

### ‚úÖ **Important Points:**

* Only **Hermitian matrices** (i.e., $A^* = A$) are **guaranteed** to be unitarily diagonalizable.
* Not all square matrices are unitarily diagonalizable.
* **Unitary diagonalization** preserves inner product and length.
* Diagonalization reveals eigenstructure clearly.

---

### ‚ùå **Misconceptions:**

* Not every matrix is unitarily diagonalizable.
* Non-Hermitian or non-normal matrices may not be diagonalizable at all.
