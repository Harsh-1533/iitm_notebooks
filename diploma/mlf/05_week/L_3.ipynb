### ðŸ”· What is a Unitary Matrix?

* A **unitary matrix** is a square matrix $U \in \mathbb{C}^{n \times n}$ such that:

$$
U^* U = I \quad \text{(where \( U^* = \overline{U}^T \) is the conjugate transpose)}
$$

* Equivalent definition:
  The **columns of a unitary matrix** form an **orthonormal basis** in $\mathbb{C}^n$.

### ðŸ”¹ Properties of Unitary Matrices

#### âœ… **Property 1: Length is Preserved**

For any vector $x \in \mathbb{C}^n$, if $U$ is unitary:

$$
\|Ux\| = \|x\|
$$

ðŸ”¹ **Proof Sketch**:

$$
\|Ux\|^2 = (Ux)^* (Ux) = x^* U^* U x = x^* I x = \|x\|^2
$$

> So unitary matrices act like **rigid rotations/reflections** in complex spaceâ€”they donâ€™t stretch or shrink vectors.

---

#### âœ… **Property 2: Eigenvalues Have Magnitude 1**

If $\lambda$ is an eigenvalue of a unitary matrix $U$, then:

$$
|\lambda| = 1
$$

> That means: **All eigenvalues lie on the unit circle in the complex plane**.

ðŸ”¹ **Proof Sketch**:
Let $Ux = \lambda x$ with $x \neq 0$:

$$
\|Ux\| = \|\lambda x\| = |\lambda| \|x\| = \|x\| \Rightarrow |\lambda| = 1
$$

---

#### âœ… **Property 3: Eigenvectors of Distinct Eigenvalues are Orthogonal**

Let:

$$
Ux = \lambda_1 x, \quad Uy = \lambda_2 y, \quad \lambda_1 \ne \lambda_2
$$

Then:

$$
x^* y = 0 \Rightarrow \text{x and y are orthogonal}
$$

> This is **shared with Hermitian matrices** as well.

ðŸ”¹ **Proof Idea**:
From $Ux = \lambda_1 x$, $Uy = \lambda_2 y$, and the unitary property, you derive:

$$
x^* y = \lambda_1^* \lambda_2 x^* y
\Rightarrow (\lambda_1^* \lambda_2 - 1)x^* y = 0
$$

Since $\lambda_1 \ne \lambda_2$, we get $x^* y = 0$.

### ðŸ§­ Unitary Matrix in Diagonalization

For any **Hermitian matrix** $A$, we can **diagonalize** it using a unitary matrix $U$:

$$
A = U \Lambda U^*
$$

* $U$: unitary matrix of eigenvectors of $A$
* $\Lambda$: diagonal matrix of real eigenvalues of $A$

ðŸ”¸ This is part of the **Spectral Theorem** for Hermitian matrices.

### ðŸ”” Why Important in ML?

Unitary matrices are essential for:

* Diagonalizing Hermitian (especially covariance) matrices.
* Quantum computing (unitary evolution).
* Stability-preserving transformations (lengths stay preserved).
* Principal Component Analysis (PCA) in complex vector spaces.

## âœ… Summary of Properties of Unitary Matrices

### ðŸ”¹ 1. **Definition of Unitary Matrix**

* A matrix $U \in \mathbb{C}^{n \times n}$ is **unitary** if:

  $$
  U^* U = U U^* = I
  $$

  where $U^*$ is the **conjugate transpose** of $U$.
* **Equivalently**:

  * It must be a **square matrix**.
  * Its **columns form an orthonormal set** (orthogonal + unit length).

---

### ðŸ”¹ 2. **Eigenvalue Property**

* If $\lambda$ is an eigenvalue of a unitary matrix, then:

  $$
  \boxed{|\lambda| = 1}
  $$

  This means all eigenvalues lie **on the unit circle** in the complex plane.

---

### ðŸ”¹ 3. **Eigenvectors of Unitary Matrices**

* **Eigenvectors corresponding to distinct eigenvalues are orthogonal**:

  $$
  \lambda_1 \ne \lambda_2 \Rightarrow v_1 \cdot v_2 = 0
  $$
* They may not be orthonormal **unless normalized**.

---

### ðŸ”¹ 4. **Unitary Matrix â‰  Hermitian Matrix**

* A **unitary matrix is not necessarily Hermitian**.

  * A **Hermitian matrix** satisfies: $A = A^*$
  * A **unitary matrix** satisfies: $U^* = U^{-1}$
* So, **every Hermitian matrix is not unitary**, and **every unitary matrix is not Hermitian**.

---

### ðŸ”¹ 5. **Equivalence Conditions**

If $U$ is unitary, the following are always true:

* $U^* U = I$
* The **columns of $U$** are an **orthonormal set**.
* $U^*$ is also unitary.
* $U^{-1} = U^*$
