## ✅ **Spectral Theorem – Summary Notes**

---

### 📌 **Main Statement (Complex Vector Space):**

> A **Hermitian matrix** is **unitarily diagonalizable**.

This means:

$$
\text{If } A = A^*, \text{ then } \exists \text{ unitary } U \text{ such that } A = U \Lambda U^*
$$

* $A^*$: Conjugate transpose of $A$
* $U$: Unitary matrix (i.e., $U^* = U^{-1}$)
* $\Lambda$: Diagonal matrix containing real eigenvalues of $A$

---

### 📌 **Real Version (Real Vector Spaces):**

> A **real symmetric matrix** is **orthogonally diagonalizable**.

That is:

$$
A = Q \Lambda Q^T
$$

* $Q$: Orthogonal matrix (i.e., $Q^T = Q^{-1}$)
* $\Lambda$: Diagonal matrix with **real eigenvalues**
* Applies when $A = A^T$

---

### ✅ **Proof Outline Using Schur’s Theorem:**

1. From **Schur’s Theorem**, any square matrix $A$ is **unitarily similar** to an **upper triangular matrix**:

   $$
   A = U T U^* \quad \text{(with } T \text{ upper triangular)}
   $$

2. If $A$ is **Hermitian** (i.e., $A = A^*$), then:

   $$
   T = T^* \Rightarrow T \text{ is also Hermitian}
   $$

3. The only matrix that is **upper triangular and Hermitian** is a **diagonal matrix**.

   * Hence, $T$ must be diagonal.
   * Therefore, $A = U \Lambda U^*$

---

### 🔍 **Important Observations:**

| Property        | Hermitian Matrix                                                           |
| --------------- | -------------------------------------------------------------------------- |
| Diagonalizable? | Yes, via **unitary matrix**                                                |
| Eigenvalues     | Always **real**                                                            |
| Eigenvectors    | Can form an **orthonormal basis**                                          |
| Converse true?  | ❌ No! A matrix **can be unitarily diagonalizable** without being Hermitian |

---

### 🚫 **Converse is False – Example:**

Consider matrix:

$$
A = \begin{bmatrix}
0 & -1 \\
1 & 0
\end{bmatrix}
$$

* $A$ is **not Hermitian** (since $A^* \neq A$)
* But it **is unitarily diagonalizable** (with eigenvalues $i$, $-i$)

---

### ✅ **Easy Example:**

Let:

$$
A = \begin{bmatrix}
1 & 1+i \\
1-i & 4
\end{bmatrix}
$$

Check:

* $A^* = A$ → So, **Hermitian**
* Eigenvalues: $\lambda_1 = 1$, $\lambda_2 = 4$
* Eigenvectors: $z_1 = [-1+i, 1]^T$, $z_2 = [1-i, 2]^T$
* Normalize eigenvectors → form unitary matrix $U$
* Then:

  $$
  U^*AU = \begin{bmatrix}
  1 & 0 \\
  0 & 4
  \end{bmatrix}
  $$

✅ Verified: Hermitian matrix is **unitarily diagonalizable** with **real eigenvalues**.

### 📚 **Key Takeaways:**

* **Spectral Theorem** is a powerful tool in linear algebra.
* Hermitian ⇒ Unitarily Diagonalizable ✅
* Real Symmetric ⇒ Orthogonally Diagonalizable ✅
* Unitarily Diagonalizable ⟹/ Hermitian ❌
* Used in quantum mechanics, PCA, and more ML techniques.

### 🔷 **Key Properties and Theorems**

#### ✅ 1. **Schur's Theorem**

* **Statement:**
  For any $n \times n$ **complex matrix** $A$, there exists a **unitary matrix** $U$ and an **upper triangular matrix** $T$ such that:

  $$
  A = U T U^*
  $$
* 🧠 This theorem is used as a foundational step in proving diagonalizability of Hermitian matrices.

---

#### ✅ 2. **Spectral Theorem**

* **Statement:**
  For any **Hermitian matrix** $A$, there exists a **unitary matrix** $U$ such that:

  $$
  D = U^* A U
  $$

  where $D$ is a **diagonal matrix with real entries**.
* 🧠 This means Hermitian matrices have **real eigenvalues** and are **unitarily diagonalizable**.

---

#### ✅ 3. **Diagonalizability Facts**

* ❌ Not every unitarily diagonalizable matrix is Hermitian.
  (E.g., $A = \begin{bmatrix} 0 & -1 \\ 1 & 0 \end{bmatrix}$ is unitarily diagonalizable but not Hermitian.)
* ✅ Every Hermitian matrix **is** unitarily diagonalizable.
  This is guaranteed by the **Spectral Theorem**.

---

### 📌 Summary of Important Results

| Property                         | Hermitian Matrix | Unitarily Diagonalizable Matrix |
| -------------------------------- | ---------------- | ------------------------------- |
| Has real eigenvalues             | ✅ Yes            | ❌ Not necessarily               |
| Diagonalizable by unitary matrix | ✅ Always         | ✅ By definition                 |
| Must be Hermitian                | ✅ True           | ❌ False                         |
| Diagonal form is real            | ✅ Always         | ❌ Not always                    |
