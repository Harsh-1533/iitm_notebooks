## ðŸ”¹ **What Is Diagonalization?**

A **matrix is diagonalizable** if it can be written in the form:

$$
A = PDP^{-1}
$$

Where:

* $A$: the original square matrix
* $D$: a **diagonal matrix** containing **eigenvalues** of $A$
* $P$: a matrix whose **columns are eigenvectors** of $A$
* $P^{-1}$: the inverse of $P$

---

## ðŸ§  **Why Diagonalize a Matrix?**

Diagonalization simplifies many matrix computations:

* Powers of matrix: $A^n = P D^n P^{-1}$
* Solving systems of differential equations
* Understanding matrix behavior (e.g., stability, transformation)
* Used in **PCA**, **quantum mechanics**, and **Markov chains**

---

## âœ… **Conditions for Diagonalizability**

A matrix $A \in \mathbb{R}^{n \times n}$ is diagonalizable **if and only if** it has **$n$ linearly independent eigenvectors**.

### Key Points:

* If $A$ has **distinct eigenvalues**, it's **always diagonalizable**.
* If $A$ is **symmetric**, it is **always diagonalizable**, and $P$ is **orthogonal**: $P^{-1} = P^T$

## ðŸ§® **How to Diagonalize a Matrix** â€“ Step-by-Step

Letâ€™s say:

$$
A = \begin{bmatrix}
4 & 1 \\
0 & 2
\end{bmatrix}
$$

### Step 1: Find eigenvalues

Solve the **characteristic equation**:

$$
\det(A - \lambda I) = 0
\Rightarrow \begin{vmatrix}
4 - \lambda & 1 \\
0 & 2 - \lambda
\end{vmatrix} = (4 - \lambda)(2 - \lambda)
$$

So, eigenvalues:

$$
\lambda_1 = 4,\quad \lambda_2 = 2
$$

---

### Step 2: Find eigenvectors

#### For $\lambda = 4$:

$$
(A - 4I) x = 0 \Rightarrow \begin{bmatrix}
0 & 1 \\
0 & -2
\end{bmatrix}
\Rightarrow \text{Eigenvector: } x_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}
$$

#### For $\lambda = 2$:

$$
(A - 2I) x = 0 \Rightarrow \begin{bmatrix}
2 & 1 \\
0 & 0
\end{bmatrix}
\Rightarrow \text{Eigenvector: } x_2 = \begin{bmatrix} -1 \\ 2 \end{bmatrix}
$$

---

### Step 3: Construct $P$ and $D$

$$
P = \begin{bmatrix} 1 & -1 \\ 0 & 2 \end{bmatrix}, \quad D = \begin{bmatrix} 4 & 0 \\ 0 & 2 \end{bmatrix}
$$

So:

$$
A = PDP^{-1}
$$
